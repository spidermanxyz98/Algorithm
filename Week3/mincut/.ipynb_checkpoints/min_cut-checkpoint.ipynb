```python
import pandas as pd
import numpy as np
from copy import deepcopy
```

```python
#label is a binary array with 1 and 0. findindex returns the index of kth
#nonzero item. k start from 1 and the returned index start from zero
def findindex(label,k):
    j=0
    for i,item in enumerate(label):
        if item != 0:
            j+=1
        if j==k:
            return(i)
    
```

```python
#graph is a list with each element an array. The ith array contains the neighbor of the ith vertex.
#labels is used to label the vertex which are still active (1 for active 0 for not inactive) 
#v1 and v2 (start from zero v1 < v2) are the index of the selected two vetices about to fuse.
def fusion(graph, labels, v1, v2):
    #eliminate vertex v2
    labels[v2]=0
    
    #remove the slef loop
    if (v2) in graph[v1]:
        index1,=np.where(graph[v1]==v2)
        index2,=np.where(graph[v2]==v1)
        graph[v2]=np.delete(graph[v2],index2)
        graph[v1]=np.delete(graph[v1],index1)
    
    graph[v1]=np.concatenate([graph[v1],graph[v2]])
    
    #Since v2+1 is fused to v1 +1, we need to update all the v2+1 with v1+1
    for i in np.arange(len(label)):
        np.place(graph[i],graph[i]==v2, v1)
            
    
    
```

```python
test=np.array([np.array([1,2,3]),np.array([0,3]),np.array([0,3]),np.array([0,1,2])])
label=np.array([1,1,1,1],dtype=np.int8)
test_cp=deepcopy(test)

fusion(test_cp,label,0,1)
print(test_cp)
print(label)
print(test)
#fusion(test, label,0,3)
#print(test)
#print(label)
```

```python
mincut=[]

for s in np.arange(10000):
    #print('*'*10)
    n=len(graph)
    labels=np.ones(n, dtype=np.int8)
    graph1=deepcopy(graph)

    for i in np.arange(n-2):
    #randomly choose the first vertex
        k1=np.random.randint(0,n-i)
        v1=findindex(labels,k1+1)
        #print('v1:',v1)
        
    #randomly choose one of v1's neighbor
        index_v2=np.random.randint(0,len(graph1[v1]))
        v2=graph1[v1][index_v2]
        #print('v2:',v2)
        
    #make sure v1 is the smaller one 
        if v1>v2:
            v1, v2 = v2, v1

        #print(v1,v2)
        fusion(graph1,labels,v1,v2)
        #print(labels)
        #print(graph1)
        #print(graph)
    mincut.append(len(graph1[0]))


#mincut_eff=list(filter(lambda x: x>0, mincut))
print(min(mincut))
print(mincut)
```
